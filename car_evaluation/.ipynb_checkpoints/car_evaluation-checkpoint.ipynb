```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df= pd.read_csv("/kaggle/car_evaluation/car_evaluation.csv")

df.head()

col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df.columns = col_names
X = df.drop('class', axis=1)
y = df['class']

#  Set up plot
fig, ax = plt.subplots(3, 3, figsize=(12,11))
ax = ax.flatten()
fig.suptitle("Value Distributions")
fig.tight_layout()

# show distribution of values in each column.
for i, col in enumerate(col_names):
    vc = df[col].value_counts()
    ax[i].bar(vc.keys(), vc.values)
    ax[i].set_xlabel(col)
    ax[i].set_ylabel("value count")

# hide unused plot
for j in range(len(col_names), len(ax)):
    fig.delaxes(ax[j])

    

```
